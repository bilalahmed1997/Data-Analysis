# LOC & ILOC

### syntax : [row , column]

label indexing ki sequence khrab hoskti

label indexing ko ankhon se dkh skte

label indexing datatype (int , flt , string)

it is front end indexing

Note : it also includes the last value

```python
# single column

f500['profits']

f500.loc[: , 'profits']

# selective column

f500.loc[: , ['profits' , 'ceo']] # fancy indexing

# range

f500.loc[: , "profits" : "ceo"]
```

```python
# single row

f500.loc[0,:]

# selective rows

f500.loc[[0,10,20],:] # fancy indexing

# range

#f500.loc[0:5]
```

# Boolean Indexing

Used For filtering

```python
f500

f500["country"]

f500["country"] == "USA"

bool_usa = f500["country"] == "USA"

bool_usa
```

```python
usa_com = f500[bool_usa]
```

```python
usa_com
```
