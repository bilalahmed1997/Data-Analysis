```python
# Convert Ram column into integer
laptops.loc[: , "RAM"]

laptops.loc[: , "RAM"] = laptops.loc[: , "RAM"].str.replace("GB" , "")
```

```python
# Convert Price into float

laptops.loc[: , "Price Euros"]

laptops.loc[: , "Price Euros"].str.replace("," , ".")

laptops.loc[: , "Price Euros"] = laptops.loc[: , "Price Euros"].str.replace("," , ".")
```

# Transformation

    Grouping / Pivoting

```python
# index === jiski base pr grouping karni hai
# values = desired aggregation column
# aggfunc = desired aggregation function

laptops.pivot_table(index = "Manufacturer" ,\
                    values = "Price Euros" , aggfunc = "sum")
```

```python
laptops.pivot_table(index = "Manufacturer" ,\
                    values = "Price Euros" , aggfunc = ["sum" , "min" , "max"])
```

# Concat

    join/merge

   combine two tables
   table 1 : top 3 most exp products of apple
   table 2 : top 3 most exp products of hp
   
   steps:
   1) filtering
   2) sort 
   3) head(3) 

```python
app_bool = laptops["Manufacturer"] == "Apple"
apple = laptops[app_bool]
```

```python
apple = apple.sort_values("Price Euros" , ascending = False).head(3)
```

```python
apple
```

```python
# For HP

app_bool = laptops["Manufacturer"] == "HP"
hp = laptops[app_bool]
hp = hp.sort_values("Price Euros" , ascending = False).head(3)
hp
```

```python
# concat

concat_df = pd.concat([apple , hp] , ignore_index = True)
concat_df
```
